```python
# Cell 1: Authenticate
import sys
sys.path.append('~/home/phoenyx/projects/labarchives-api/src/labapi')

import notebook_logger

# Replace with your LabArchives notebook name
logger = NotebookLogger(notebook_name="My Research")
```

```python
# Cell 2: Computation + plot
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("sin(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.tight_layout()
plt.show()

# The last expression becomes the logged result
np.trapz(y, x)
```

```python
# Cell 3: Save UI
# Enter tags (comma-separated) and click "Save to LabArchives"
logger.show_ui()
```
